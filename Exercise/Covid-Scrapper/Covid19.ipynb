```python
import requests
import bs4
```

```python
URL = "https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory"
html = requests.get(URL).content
```

```python
soup = bs4.BeautifulSoup(html, "html.parser")
```

```python
covid19 = soup.find("div", id="covid19-container").tbody.tr.find_next("tr")
countries = []
for covid19 in covid19.find_all_next("tr"):
    country = covid19.a.get_text().strip()
    if country == "UTC": break
    row = {"name": country}
    row_names = ["cases", "deaths", "recoveries"]
    for name, ele in zip(row_names, covid19.find_all("td")):
        if ele.get_text() == "No data\n":
            value = None
        else:
            value = int(ele.get_text().strip().replace(",", ""))
        row[name] = value
    countries.append(row)
```

```python
import json

print(json.dumps(countries, indent=2))
```
